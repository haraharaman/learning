### ノック１３：データに揺れがあるまま集計しよう

```python
sales["purchase_date"] = pd.to_datetime(sales["purchase_date"])
sales["purchase_month"] = sales["purchase_date"].dt.strftime("%Y%m")
```

```python
pv = sales.pivot_table(index="purchase_month", columns="item_name", aggfunc="size",fill_value=0)
pv
```

### ノック１４：商品名の揺れを補正しよう

```python
sales["item_name"] = sales["item_name"].str.upper()
sales["item_name"] = sales["item_name"].str.replace(" ","")
sales["item_name"] = sales["item_name"].str.replace("　","")
```

```python
pd.unique(sales["item_name"])
pv = sales.pivot_table(index="purchase_month" ,columns="item_name", aggfunc="size",fill_value=0)
pv
```

### ノック１５：金額欠損値の補完をしよう

> インデントされたブロック



```python
sales
```

```python
nan_list = sales["item_price"].isnull()
for product in sales.loc[nan_list,"item_name"].unique():
  sales.loc[(nan_list)&(sales["item_name"]==product),"item_price"] = sales.loc[(~nan_list) & (sales["item_name"] == product), "item_price"].max()
sales.isnull().any() #要素にnanがあればtrueを返す
```

```python
sales
```

### ノック１６：顧客名の揺れを補正しよう

```python
customer["顧客名"]=customer["顧客名"].str.replace("　","")
customer["顧客名"]=customer["顧客名"].str.replace(" ","")
```

```python
customer
```

### ノック１７：日付の揺れを補正しよう

```python
#文字列になっている行番号の抽出
datelist = customer["登録日"].astype("str").str.isdigit()

tmp1 = pd.to_datetime(customer.loc[(~datelist),"登録日"])
tmp2 = pd.to_timedelta(customer.loc[(datelist), "登録日"].astype("float"), unit="D") + pd.to_datetime("1900/01/01")
pd.concat([tmp1,tmp2])
```

```python
customer["登録日（加工後）"] = pd.concat([tmp1,tmp2])
customer
```

### ノック１８：顧客名をキーに２つのデータを結合(ジョイン)しよう

```python
data = pd.merge(sales,customer,left_on="customer_name",right_on="顧客名",how = "left")
```

### ノック１９：クレンジングしたデータをダンプしよう

```python
data.drop(["customer_name","登録日"],axis=1)
#表示上消えるのみ
```
